```r
par(family='AppleGothic',mfrow=c(2,2))

# 코로나 시기(2020-2021) 추출
코로나 <- townD[41:80,]
# arrange(코로나,년도,관광특구)
name = c("20-1","20-2","20-3","20-4","21-1","21-2","21-3","21-4")

# 각 그래프 변수 지정
이태원점포당매출 <- subset(코로나,관광특구=="이태원")$점포당매출금액
명동점포당매출 <- subset(코로나,관광특구=="명동 남대문 북창동 다동 무교동")$점포당매출금액
동대문점포당매출 <- subset(코로나,관광특구=="동대문패션타운")$점포당매출금액
종로청계점포당매출 <- subset(코로나,관광특구=="종로 청계")$점포당매출금액
잠실점포당매출 <- subset(코로나,관광특구=="잠실")$점포당매출금액


plot(이태원점포당매출/1000000,
     type="o",
     ann = F, # x,y의 label 제거
     axes = F, # x,y좌표 frame 제거
     ylim=c(0,150),
     )

lines(명동점포당매출/1000000,
     type = "o",
     col = "red",
     lty = 1
     )
lines(동대문점포당매출/1000000,
     type = "o",
     col = "blue",
     lty = 1
     )
lines(종로청계점포당매출/1000000,
     type = "o",
     col = "green",
     lty = 1
     )
lines(잠실점포당매출/1000000,
     type = "o",
     col = "purple",
     lty = 1
     )
# 선긋기
segments(6,0,6,100,
     col='red',
     lty=2
)
# 글자 넣기
text(6,120, '특이점', col='dark red',srt=90) # srt : 회전
# 선긋기
segments(8,0,8,120,
     col='red',
     lty=2
)
# 글자 넣기
text(8,140, '특이점', col='dark red',srt=90) # srt : 회전


title(main="2020-2021 분기/관광특구별 점포당매출") # main title
title(ylab = "점포당매출금액(단위:백만 원)", col.lab = "darkblue") # y축 (label) 지정
axis(1,at=1:8,lab=name,cex.axis=0.5) # x축 lim
axis(2,cex.axis=0.7) # y축 lim
legend(3, 150, # legend 박스 위치 좌표
    c('이태원','명동','동대문','종로','잠실'), # 각 그래프의 이름
    col = c('black','red','blue','green','purple'), # 각 그래프의 색상
    #pch=c(21,22,23), # 각 그래프의 점 종류
    lty=1, # 각 그래프의 선 종류
    cex=0.5, # 박스 사이즈 줄이기(default=1)
) 

# 21년 2분기에 왜 증가했을까?

### 분석 1 : 확진자 수와의 연관성

# 각 그래프 변수 지정
이태원확진 <- subset(코로나,관광특구 =="이태원")$점포당매출금액대비확진자수
명동확진 <- subset(코로나,관광특구 =="명동 남대문 북창동 다동 무교동")$점포당매출금액대비확진자수
동대문확진 <- subset(코로나,관광특구 =="동대문패션타운")$점포당매출금액대비확진자수
종로청계확진 <- subset(코로나,관광특구 =="종로 청계")$점포당매출금액대비확진자수
잠실확진 <- subset(코로나,관광특구 =="잠실")$점포당매출금액대비확진자수

plot(이태원확진,
     type="o",
     ann = F, # x,y의 label 제거
     axes = F, # x,y좌표 frame 제거
     ylim=c(0,40000),
       )

lines(명동확진,
     type = "o",
     col = "red",
     lty = 1
     )
lines(동대문확진,
     type = "o",
     col = "blue",
     lty = 1
     )
lines(종로청계확진,
     type = "o",
     col = "green",
     lty = 1
     )
lines(잠실확진,
     type = "o",
     col = "purple",
     lty = 1
     )

title(main="2020-2021 분기/관광특구별 확진자수") # main title
title(ylab = "확진자수(단위:명)", col.lab = "darkblue") # y축 (label) 지정
axis(1,at=1:8,lab=name,cex.axis=0.5) # x축 lim
axis(2, cex.axis=0.7) # y축 lim
legend(1, 40000, # legend 박스 위치 좌표
    c('이태원','명동','동대문','종로','잠실'), # 각 그래프의 이름
    col = c('black','red','blue','green','purple'), # 각 그래프의 색상
    #pch=c(21,22,23), # 각 그래프의 점 종류
    lty=1, # 각 그래프의 선 종류
    cex=0.5, # 박스 사이즈 줄이기(default=1)
) 
# 결과 : 점포당매출금액대비확진자수와 점포당매출금액은 유의미한 상관관계가 없다.
plot(이태원확진,이태원점포당매출/1000000,
     main='2020-2021 분기별 '
     ylab='점포당매출(단위:백만 원)',
     xlab='확진자수(단위:명)',
     )
plot(명동확진,명동점포당매출/1000000,
     ylab='점포당매출(단위:백만 원)',
     xlab='확진자수(단위:명)'
     )
plot(동대문확진,동대문점포당매출/1000000,
     ylab='점포당매출(단위:백만 원)',
     xlab='확진자수(단위:명)'
     )
plot(종로청계확진,종로청계점포당매출/1000000,
     ylab='점포당매출(단위:백만 원)',
     xlab='확진자수(단위:명)'
     )
plot(잠실확진,잠실점포당매출/1000000,
     ylab='점포당매출(단위:백만 원)',
     xlab='확진자수(단위:명)'
     )
```
