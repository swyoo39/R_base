```r
x <- 10
y <- runif(1)*100
if(x>y){
    print("x>runif()")
}
print("--- End ---")
```

```r
x <- 10
y <- runif(1)*100
if(x>y){
    print("x>runif()")
}else {
    print("x<runif()")
}
print("--- End ---")
```

```r
x <- 10
y <- runif(1)*100
if(x>y){
    print("x>runif()")
}else if(x<y){
    print("x<runif()")
}else {
    print("x=runif()")
}
print("--- End ---")
```

```r
# 홀수 짝수 구분
a <- 1234 # vector일 경우 정확한 판단을 하지 못한다.
if (a%%2==0){
    cat(a,": 짝수")
    sprintf("%i : 짝수",a)
}else{
    cat(a,": 홀수")
}
```

```r
a <- c(8,8,7)
ifelse(a%%2 == 0, paste(a," : 짝수"), paste(a," : 홀수"))
paste(a,ifelse(a%%2 == 0, " : 짝수", " : 홀수"))
```

```r
# 사용자로부터 data를 입력받아 숫자가 홀/짝인지 판단하여 출력하기
num1 <- as.integer(readline("숫자를 입력하세요 :"))

# 입력하신 숫자는 12이고 짝수입니다.
ifelse(num1%%2 == 0, 
       paste('입력하신 수는 ',num1,"이고 짝수입니다."), 
       paste('입력하신 수는 ',num1,"이고 홀수입니다."))
paste('입력하신 수는',num1,'이고',ifelse(num1%%2==0,"짝수","홀수"),'입니다.')
```

```r
n1 <- as.integer(readline("점수를 입력하세요 : "))
paste(ifelse(n1>=90, "A학점", 
             ifelse(n1>=80, "B학점", 
                    ifelse(n1>=70, "C학점"))), 
      '입니다.')
```
