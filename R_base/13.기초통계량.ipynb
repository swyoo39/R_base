# 기본 통계량으로 분석하기


```r
# 밀가루 복용
flour <- c(3, -2, -1, 0, 1, -2)

# 다이어트 약 복용
diet <- c(-4, 1, -3, -5, -2, -8)
```

```r
total <- c(flour,diet)
total
```

```r
# 기본 통계량
summary(total)
total
```

```r
# histogram
par(bg='white',mfrow=c(1,2))
hist(total)
boxplot(flour,diet)
```

```r
# 상관계수
cor(flour,diet) # -0.0170805051030557 : 상관관계가 없다.
```

```r
x <- c( 1, 9, 3) # 벡터의 순서가 바뀌면 상관계수도 바뀌므로 주의
y <- c(10,20,30)
```

```r
par(mfrow=c(1,2),bg='white')
plot(x,type='l')
plot(y,type='l')
```

```r
cor(x,y)
```

```r
# 커피판매량 빈도수
table(cafe$Coffee)
```

```r
# 범위 빈도수
coffee_range <- seq(0,60,10) # 범위
table(cut(cafe$Coffee, breaks = coffee_range,right=F)) 
# cut(자른다, Coffee를, coffee_range 기준으로), right=F : (오른쪽)끝 값 포함 안 함
```

```r
# 줄기-잎 그림
stem(cafe$Coffee) # 10의 자리 | 1의자리 나열(개수)
```
