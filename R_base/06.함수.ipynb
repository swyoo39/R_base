```r
# A 컬럼의 최대,최소,(값의)빈도 구하기
max(test$A)
min(test$A)
print(table(test$A))
```

```r
# 각 컬럼의 빈도수 구하기
ncol(test) # col 수
table(test[,1]) # table : dataframe의 빈도를 보여준다.
colnames(test)[1] # 1열 이름
for(i in 1:ncol(test)){
    cat(colnames(test)[i],"컬럼의 빈도분석 결과\n")
    print(table(test[,i]))
    cat("---------------------\n")
}
```

```r
# 각 컬럼 빈도의 최대,최소 구하기
ncol(test) # col 수
colnames(test)[1] # 1열 이름
max(table(test[1])) # test 1행 빈도의 최대
for(i in 1:ncol(test)){
    cat(colnames(test)[i],"컬럼 빈도의 최대값과 최소값\n")
    x <- table(test[,i])
    cat('최대값 :',max(x),', 최소값 :',min(x),'\n')
    cat("-----------------------\n")
}
```

```r
# 각 컬럼의 빈도수 구하기와 각 컬럼의 빈도수 최대값,최소값 구하는 함수 만들기
tableMaxMin <- function(x){
    for(i in 1:ncol(x)){
        cat(colnames(x)[i],"컬럼의 빈도분석 결과")
        print(table(x[,i]))
        cat('\n')
        y <- table(x[,i])
        cat(colnames(test)[i],"컬럼 빈도의 최대값과 최소값\n")
        cat('최대값 :',max(y),', 최소값 :',min(y),'\n\n')
        cat("---------------------\n\n")
    }
}
```

```r
tableMaxMin(test)
```
