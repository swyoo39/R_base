# 주성분 분석
- 서로 연관된 변수들이 관측되었을 때, 이 변수들이 전체적으로 가지고 있는 정보를 최대한 확보하는 적은 수의 새로운 변수를 생성하는 방법
- 차원을 줄이는 방법으로, 성능 향상과 컬럼들의 결합에 목적이 있음.

```r
data <- read.table('../data/pasta.txt')
```

```r
head(data)
```

```r
# 다중 공선성 확인 : 너무 상관계수가 높아서 마치 쌍둥이인 데이터 찾기
cor(data)
```

```r
sd(data1[,3])
apply(data1,2,sd) # 각각 연산해서 한번에 보여주기 
```

# USArrests
- 1973년 미국 50개 주에서 각각 10만 명당 폭행,살인,강간으로 체포된 통계
- 또한 도시지역에 살고있는 인구의 비율도 주어진다.

```r
head(USArrests)
```

- Murder : 살인
- Assault : 폭행
- UrbanPop : 도시 인구밀도
- Rape : 강간

```r
# 데이터 상관 관계
cor(USArrests)
# 도시 인구는 살인과 폭행은 상관이 없다. 강간과는 약한 상관관계가 있다.
```

```r
# 주성분 분석
result <- princomp(USArrests,cor=T)
summary(result)
```

```r
biplot(result, scale=T,
  cex=0.5
)
# 화살표와 멀수록 안전한 지역이다.
```

```r
result$loadings
```

### Comp.1
> 1주성분 = 0.536 * Murder + 0.583 * Assault + ... <br>
> y1 = a1 * x1 + a2* x2 + ...

```r
# 1주성분으로 새로운 변수값 생성
score <- 0.536*USArrests$Murder + 0.583*USArrests$Assault + 0.278*USArrests$UrbanPop + 0.543*USArrests$Rape
head(score)
```

```r
Result1 <- data.frame(States=rownames(USArrests),Score=score)
head(Result1)
```
