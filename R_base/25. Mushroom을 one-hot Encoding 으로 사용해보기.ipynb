# column 이름과 one-hot encoding 한번에 처리하기

```r
mr3 <- data.frame(V1=mr$V1)
head(mr3)
```

```r
count = 2  

for(i in 2:23){
    for(j in sort(unique(mr[,i]))){
        mr3 = cbind(mr3,mr[,1]==j)
        colnames(mr3)[count]<- paste("V",i,"_",j,sep='')
        count= count+1
    }
}
head(mr3)
```

```r
library(nnet);library(party);library(randomForest);library(e1071);library(caret)
```

```r
mr3 <- arrange(mr3,V1)
```

```r
set.seed(1234)
samp <- createDataPartition(mr3$V1,p=0.7,list=F)
data.tr <- mr3[samp,]
data.test<- mr3[-samp,]
x <- subset(data.test, select=-V1)
y <- data.test$V1
```

```r
model.rf <- randomForest(V1~.,data=data.tr,ntree=100)
pred <- predict(model.rf,x)
mean(pred==y)
table(pred,y)
```
