```r
# maint title : Simple Time Series
y <- c(1,3,2,4,8,5,7,6,9)
```

```r
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1), # y는 벡터이므로 min, max(length) 활용
    
)
```

```r
# 꺾은선 그래프

plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1),
    type = 'l' # 꺾은 선 타입
)
```

```r
# 점과 꺾은선 그래프
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1),
    type = 'b' # 꺾은 선 타입과 점을 모두 표현
    
)
```

```r
# 점이 빠진 꺾은선 그래프
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1), 
    type = 'c' # 점이 빠진 꺾은선 그래프
    
)
```

```r
# 점과 꺾은선을 중첩한 그래프
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1), # y는 벡터이므로 min, max(length) 활용
    type = 'o', # overlap
    col = c(rep('red',3),'blue',rep('green',6))
    
)
```

```r
# 수직선 그래프
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1),
    type = 'h', # 수직선
    col = c(rep('red',3),'blue',rep('green',6))
    
)
```

```r
# 왼쪽 값을 기초로한 계단 모양 그래프
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1), 
    type = 's', # 왼쪽 계단형, step
    
)
```

```r
# 오른쪽 값을 기초로한 계단모양 그래프
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1), 
    type = 'S', # 오른쪽 계단형
    col = c(rep('red',3),'blue',rep('green',6))
    
)
```

```r
# 선 모양 실선, 점선 등등
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1), 
    type = 'S', # 오른쪽 계단형
    col = c(rep('red',3),'blue',rep('green',6)),
    lty = 6 # 6 이상 입력하면 6으로 나눈 나머지로 적용, 각 값마다 선의 모양이 다름.
)
```

```r
# linetype
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1), 
    type = 'S', # 수직선
    col = c(rep('red',3),'blue',rep('green',6)),
    lty = 6 # 6 이상은 6으로 나눈 나머지로 취급
)
```

```r
# 그래프 옵션, Plotting Symbol
plot(y,
    main = 'Simple Time Series',
    ylim = c(min(y)-1,max(y)+1),
    type = 'p', 
    col = c('red','blue','yellow'),
    pch = c(24, 20, 9) # 6 이상은 6으로 나눈 나머지로 취급, if 가능, 벡터로 해당 기호모양 출력
    
)
```

```r
x <- runif(100)
y <- runif(100)

# pch : y > 0.5 이면 1, 아니면 18
# col : y > 0.5 이면 'red', 아니면 'blue'
plot(x,y,
    col = ifelse(y>0.5, 'red','blue'),
    pch = ifelse(y>0.5, 1, 18)
)
```

### 선 그래프

```r
# 3종류 data set의 선 그래프
abc <- c(260,300,250,280,310)
def <- c(180,200,210,190,170)
ghi <- c(210,250,260,210,270)
```

```r
par(family="AppleGothic")
options(warn=-1)
plot(abc,
    type = "o", # overlap, 꼭짓점을 선 위에 표시
    col = "dark blue",
    pch=22,
    ylim = c(0,400), # 안 해주면 최소값과 최대값이 범위로 지정됨
    axes = F, # x,y 축 좌표 frame을 제거
    ann = F, # x,y 축 label을 제거
)
### 따로 빼서 쓸 경우 각각 컬러를 입힐 수 있다.

# x축과 x축의 라벨을 설정하기
axis(1, at=1:5, lab=c('서울','대전','대구','광주','원주')) 
# x축, 범위는 1:5, label 이름(범위와 대응)

# y축과 y축의 라벨을 설정하기
axis(2, ylim=c(0,400)) # y축, 범위만 0:400

# main title 지정
title(main='과일 판매량', col.main='darkred')

# x축 title(label) 지정
title(xlab='지역', col.lab='grey')

# y축 title(label) 지정
title(ylab='판매량', col.lab='grey')

# 선그리기 함수로 def 그리기, lines : 선만 그려지는 그래프
lines(def,
    type="o",
    pch=22,
    col='dark green',
    lty=2, # linetype : 점선
) 

# 선그리기 함수로 ghi 그리기
lines(ghi,
    type='o',
    pch=23,
    col='yellow',
    lty=3
)

# 범례 : legend
legend(4, 400, # legend 박스 위치 좌표
    c('야구장','축구장','해변가'), # 각 그래프의 이름
    col = c('dark blue','dark green','yellow'), # 각 그래프의 색상
    pch=c(21,22,23), # 각 그래프의 점 종류
    lty=c(1,2,3), # 각 그래프의 선 종류
    cex=0.8, # 박스 사이즈 줄이기(default=1)
) 
```

### 3D Pie Chart

```r
library(plotrix)
```

```r
sales <- c(210,110,400,550,600,130)
week <- c('월','화','수','목','금')
ratio <- sales/sum(sales)*100
label <- paste(week,'\n',ratio,'%')

par(family='AppleGothic',bg='white')
pie3D(sales,
    main='영업실적 파이차트',
    labels=label,
    #col = rainbow(length(sales)) : default color가 rainbow
    explode=0.05, # 쪼개버리기
    labelcex=0.8 # 글자 크기 80%
)
```
