### 1880년~2014년 미국의 아기 이름을 정리해놓은 데이터셋

```r
baby <- read.csv('../data/R_Baby_Names_1880-2014.csv',stringsAsFactors=F)
head(baby)
tail(baby)
```

```r
str(baby)
# name이 factor로 들어가있네?
```

```r
baby$Gender <- as.factor(baby$Gender)
```

```r
summary(baby)
```

```r
baby$Year <- substr(baby$Year,7,10)
head(baby)
```

```r
# 저장하기
write.csv(baby, '../data/baby.csv',row.names=F)
# 용량이 70mb > 50mb로 줄어든다
```

```r
summary(baby)
# Gender F가 많은 이유, 여자 이름의 종류가 더 많아서.(남자 이름은 중복값이 많다)
# 이름 수의 분포가 치우쳐있다(mean - median 차이가 크다)
```

```r
# x: year y: count
# baby에서 여자아이만 Female이라는 dataset으로 만들고 nrow 구하기
Female <- subset(baby,Gender=='F')
```

```r
nrow(Female)
```

```r
# 여자 아이 이름 중복되지 않는 갯수 구하기
length(unique(Female$Name))
```

```r
# 중복되지 않게 여자아이 이름 20개 출력하기
head(unique(Female$Name),20)
```

```r
# 가장 많이 사용된 여자아이 이름 10개 보여주기
table(Female$Name)
```

```r
head(Female)
summary(Female)
```

```r
library(dplyr) # arrange 쓰기 위함
```

```r
FemaleN <- aggregate(Count~Name,Female,sum)
head(arrange(FemaleN,desc(Count)),10)
```

```r
# baby에서 남자아이만 Male로 뽑기
Male <- subset(baby,Gender=='M')
```

```r
nrow(Male)
```

```r
# 남/여의 연도별 합계
FemaleCbyY <- aggregate(Count~Year,Female,sum)
MaleCbyY <- aggregate(Count~Year,Male,sum)
```

```r
options(warn=-1,scipen=100)
par(family="AppleGothic",bg="white")

plot(FemaleCbyY,
    main='1880~2014년 인구분포 현황',
    type='l',
    ylab='Count',
    xlab='Year',
    #cex.axis = 0.1,
    xlim=c(1880,2020),
    ylim=c(0,2500000),
    lty = 1,
    col='orange',
    #axes =F, #x,y 좌표 frame 제거
    #ann=F, #x,y label 제거
)

#axis(1, at=1:8, lab=c(1880:2020), cex.axis=0.3)
#axis(2, cex.axis=0.5)

lines(MaleCbyY,
    col='blue'
)

legend(1980,1000000,
    c("Female","Male"),
    col=c("orange","blue"),
    #pch=c(4,3),
    lty=c(1,1),
    cex=0.8
)
```
