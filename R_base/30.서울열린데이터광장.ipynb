```r
colnames(df)[1] <- '년도'
colnames(df)[2] <- '자치구명'
colnames(df)[4:ncol(df)] <- paste(1:12,'월',sep='')
```

```r
head(df)
```

```r
# 2011, 종로구, 발생건수, 1월부터 12월까지 
a <- c()
for(i in 1:length(unique(df$년도))){
  a <- rbind(a, subset(df, 자치구명=='종로구' & 구분=='발생건수')[i,4:ncol(df)]) # i는 년도 순서
}
rownames(a) <- c(2011:2019)
a
str(a)
```

```r
nrow(seoul) == length(unique(df$자치구명))*ncol(a)*nrow(a)
seoul_df <- data.frame(년도='')
str(seoul_df)

for(i in 2011:2019){
  seoul_df <- rbind(seoul_df, data.frame(년도=as.factor(rep(i,length(unique(df$자치구명))*ncol(a)))))
} 
str(seoul_df)
class(seoul_df)
```

```r
seoul_df <- seoul_df[-1,] # 빈값 없애주기
```

```r
seoul_df <- data.frame(년도=seoul_df)
```
