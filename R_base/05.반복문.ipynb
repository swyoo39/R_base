```r
str(excel)
```

```r
# excel에서 q3 컬럼 데이터의 홀수 합과 짝수 합 구하기
# 홀수의 합은 604 이고 짝수의 합은 852 입니다
q3 <- excel$q3
sumEven <- 0; sumOdd <- 0;
for(i in q3){
    if(i%%2==0){
        sumEven = sumEven + i
    }else{
        sumOdd = sumOdd + i
    }
}
cat('홀수의 합은',sumOdd,'이고 짝수의 합은',sumEven,'입니다.')
#sum(q3[seq(1,length(q3),2)])
```

```r
# excel$q3 컬럼 데이터 중 3의 배수 합 구하기
sum3 <- 0;
for(i in q3){
    if(i%%3==0){
        sum3 = sum3 + i
    }
}
cat(sum3)
```

```r
# q3 데이터 중 3의 배수도 되고 4의 배수도 되는 수의 합을 구하기
sum34 <- 0;
for(i in q3){
    if(i%%3==0 && i%%4==0){
        sum34 = sum34 + i
    }
}
cat(sum34)
```

```r
# 구구단 2단 출력하기
for(i in 1:9){
    for(j in 1:9){
        cat(i, 'X', j, '=', i*j,'\t')
    }
    cat('\n')
}
```

```r
input <- as.numeric(readline('구구단 숫자를 입력하세요 : '))
for(j in 1:9){
    cat(input, 'X', j, '=', i*j,'\t')
}
cat('\n')
```

```r
# 구구단 함수로 만들기 01
gugu <- function(){
    dan <- as.numeric(readline('구구단 숫자를 입력하세요 : '))
for(j in 1:9){
    cat(dan, 'X', j, '=', i*j,'\t')
}
cat('\n')
}
```

```r
gugu()
```

```r
# 구구단 함수 만들기 02, 매번 단 입력하기 귀찮다!
gugu <- function(x){
    for(j in 1:9){
        cat(x, 'X', j, '=', i*j,'\n')
    }
}
```

```r
gugu(2)
```

```r
# 구구단 함수 만들기 03
gugu <- function(x,y){
    for(j in 1:y){
        cat(x, 'X', j, '=', i*j,'\n')
    }
}
```

```r
gugu(2,12)
```

```r
# 2개의 벡터 자료를 함께 출력하기
name <- c('유비','관우','장비')
score <- c(85,95,89)
for(i in 1:length(score)){
    cat(name[i],':',score[i],'\n')
}
```

```r
for(i in score){
    cat(name[which(score==i)],':', i, '\n')
}
```

```r
j<-1
for(i in score){
    cat(name[j],':',i,'\n')
    j=j+1
}

# 벡터로 paste하여 한 번에 출력하기
cat(paste(name,":",score,'\n'),sep='')
```

### while

```r
i <- 0
while(i<10){
    i = i +1
    print(i)
}
```

### repeat

```r
# 괄호가 없다. 언제 끝나는지 정해줘야 함.
i <- 0
repeat{
    i = i +1
    if(i>10){
        break
    }
    print(i)
}
```
