```r
Gangnam <- c('강서구','양천구','구로구','영등포구','동작구',
             '금천구','관악구','서초구','강남구','송파구','강동구')
length(Gangnam)
Gangbuk <- c('은평구','마포구','서대문구','종로구','중구','용산구','강북구','성북구',
             '동대문구','성동구','도봉구','노원구','중랑구','광진구')
length(Gangbuk)
```

```r
library(dplyr)
```

```r
seoul$권역 <- ifelse(seoul$자치구명 %in% Gangbuk,'강북','강남') # 포함 조건 
summary(seoul)
table(seoul$권역)
seoul_Gangbuk <- seoul[seoul$권역=='강북',]
seoul_Gangbuk
seoul_Gangnam <- seoul[seoul$권역=='강남',]
seoul_Gangnam
```

```r
seoulACC.mean <- aggregate(발생건수~권역,seoul,mean)
seoulACC.sd <- aggregate(발생건수~권역,seoul,sd)
seoulACC.cv <- seoulACC.sd[2]/seoulACC.mean[2]
seoulACC.mean
seoulACC.sd
seoulACC.cv <- data.frame(권역=seoulACC.mean[1],발생건변동계수=seoulACC.cv$발생건수)
seoulACC.cv
```

```r
options(scipen = 100)
options(warn = -1)
par(family="AppleGothic")
barplot(seoulACC.cv$발생건변동계수,
        main = "권역별 발생건변동계수",
        xlab = "권역별",
        ylab = "변동계수",
       names = seoulACC.cv$권역)
```

```r
seoulInjure.mean <- aggregate(부상자수~권역,seoul,mean)
seoulInjure.sd <- aggregate(부상자수~권역,seoul,sd)
seoulInjure.cv <- seoulInjure.sd[2]/seoulInjure.mean[2]
seoulInjure.mean
seoulInjure.sd
seoulInjure.cv <- data.frame(권역=seoulInjure.mean[1],부상자변동계수=seoulInjure.cv$부상자수)
seoulInjure.cv
```
