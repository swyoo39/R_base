# Machine Learning

### 인공신경망
- 인공신경망은 Python에서 Deep Learning으로 발전됨

```r
head(iris)
```

```r
# 학습용 데이터와 검증용 데이터의 분리
set.seed(1234)
samp <- c(sample(1:50, 35) , sample(51:100,35) , sample(101:150,35))
head(samp)
```

```r
# 학습용 데이터셋
data.training <- iris[samp,]
head(data.training)
```

```r
# 검증용 데이터셋
data.test <- iris[-samp,]
```

```r
summary(data.training)
summary(data.test)
# Species 별로 잘 sampling 됐는지 확인
```

### 학습 시키기

```r
# 인공신경망 모델
library(nnet)
```

```r
# Model 만들기
model.nnet <- nnet(Species~.,data=data.training, size=3) 
# size : 은닉층 갯수, sample 수보다 작아야 함 (Hyper Parameter, 모델 외부에서 사용자가 정해주는 값)
```

```r
# 문제 및 정답 만들기 (공부 끝났으면 시험 봐야지)
x <- subset(data.test, select=-Species) # 문제
head(x)
y <- data.test$Species # 정답
head(y)
```

```r
# 검증하기
pred <- predict(model.nnet, x, type='class') # class : 분류된 것(객관식), raw : 주관식
head(pred)
```

```r
# 정확도
mean(pred == y)
# 어떻게 정확한지가 중요하다.
```

```r
# 정오표 확인 cross validation
table(pred,y)
# 선택지 간 비율이 동등해야 의미가 있는 정확도(sample 수)
```

```r
set.seed(1234)
samp<- c(sample(1:50,35), sample(51:100,35), sample(101:150,35))
```

```r
head(samp)
tail(samp)
summary(samp)
```

```r
# 학습용 데이터셋
data.tr <- iris2[samp,] # sampling 한 행들 집어넣기
summary(data.tr)
```

```r
# 테스트용 데이터셋
data.test <- iris2[-samp,]; summary(data.test)
```

```r
model.nnet <- nnet(Species ~ ., data=data.tr, size=3) # hyper parameter = 3
```

```r
# 문제 및 정답 만들기
x <- subset(data.test, select=-Species)
y <- data.test$Species
```

```r
# Test
pred <- predict(model.nnet, x, type='class')
head(pred)
```

```r
# 예측력 확인
mean(pred == y)
```

```r
# 정오표 확인
table(pred,y)
```
