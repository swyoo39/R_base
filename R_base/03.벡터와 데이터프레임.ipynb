### 두개 이상의 데이터 프레임 병합하기

```r
height <- data.frame(id=1:3, height=c(160,170,180)); height
weight <- data.frame(id=c(2,1,3), weight=c(70,60,80)); weight
```

```r
# data 수가 같고 컬럼 이름이 같을 때 병합하기
merge(height, weight)
```

```r
# key값이 되는 컬럼(이름)이 다르고, data 수도 다를 때 병합하기
height1 <- data.frame(id=1:3, height=c(160,170,180)); height1
weight1 <- data.frame(ad=c(2,1), weight=c(70,60)); weight1
```

```r
merge(height1,weight1, id=ad) # 그냥 merge하면 중복되어 섞임
```

```r
# key를 정의하기 : data가 적은 쪽의 기준으로 merge, 일종의 교집합(default값)
merge(x=height1, y=weight1, by.x='id', by.y='ad')
```

```r
# key를 정의하기 : data가 많은 쪽으로 merge, 일종의 합집합
merge(x=height1, y=weight1, by.x='id', by.y='ad', all.x=T)
```

# 문자열 처리
수집한 데이터의 필요한 문자열을 자르고, 교체하고, 추출하는 작업

### stringr package 사용
- str_length() : 문자열 길이
- str_c() : contact?, 문자열 연결
- str_sub() : 범위에 해당하는 문자열 생성
- str_split() : 구분자를 기준으로 문자열을 분리하여 부분 생성
- str_replace() : 문자열에서 기존 문자열을 특정 문자열로 변환, '123,456' >> '123456'
- str_extract() : 문자열에서 '특정 문자열 패턴'의 첫번째 '문자열' 추출
- str_locate() : 문자열에서 '특정 문자열 패턴'의 첫번째 '위치' 추출

```r
# 패키지 불러오기
library(stringr)
```

```r
string <- "hongkildong105yoube1002kwanwoo25장비2005"
```

```r
# a-z까지의 영문자가 연속으로 3개 발견된(포함된) 문자열 추출
class(str_extract_all(string, "[a-z]{3}")) # list 형태, key값 1
str_extract_all(string, "[a-z]{3}") # "정규식 패턴"에 해당하는 문자열 추출
```

```r
# a-z까지의 영문자가 연속으로 3개-5개 발견된 문자열 추출
str_extract_all(string, "[a-z]{3,5}") # "{min,max}"
```

```r
# a-z까지의 영문자가 연속으로 1개 이상 발견된 문자열 추출
str_extract_all(string, "[a-z]{1,}") # "{min,all}"
aaa <- str_extract_all(string, "[a-z]{1,}")
class(aaa)
class(unlist(aaa))
bbb <- unlist(aaa) # unlist로 vector로 바꾸기
bbb[1]
```

```r
# string에서 숫자만 출력하여 합계 구하기
sum(as.numeric(unlist(str_extract_all(string, "[0-9]{1,}")))) 
# {1,} 적지 않으면 1글자씩 추출됨
```

```r
# 한글만 추출하기
str_extract_all(string, "[가-힣]{1,}")
```

```r
# 영문자를 제외한 문자 출력하기
str_extract_all(string, "[^a-z]{1,}")
```

```r
# 문자열 길이 구하기
str_length(string)
```

```r
# 문자열 위치 구하기, 데이터마다 다른 위치를 파악하기 위함
str_locate(string, "장비") # 글자 수에 따라 start end 값의 차이가 난다.
class(str_locate(string, "장비"))
str_locate(string, "장비")[1]
```

```r
# 부분 문자열 추출
str_sub(string, 1, 32) # 1~32 까지의 글자를 추출
str_sub(string, 1, str_locate(string, "장비")[1]-1) # 1부터 "장비" 앞까지 추출
```

```r
# 대문자, 소문자로 변경하기
str_to_upper(string)
str_to_lower(string)
```

```r
# 문자열 교체하기
string <- str_replace(string, "hongkildong105", "홍길동35,")
string <- str_replace(string, "yoube1002", "유비45,")
string <- str_replace(string, "kwanwoo25", "관우25,")
string <- str_replace(string, "2005", "20")
string
```

```r
# 문자열 분리하기
str_split(string,",")
```

```r
# 천단위 쉼표가 들어간 문자열을 숫자로 변경하기
price <- "123,456,789"
str_replace_all(price,',','')
class(as.numeric(str_remove_all(price,',')))
```

```r
# NA를 변경하기
temp <- c(1,2,NA,4); temp
as.numeric(str_replace_na(temp, 10))
```
