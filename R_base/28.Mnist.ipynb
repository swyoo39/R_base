### 한 글자를 28 * 28의 pixel(행렬)로 만들고, 각각의 진하기를 255까지 정해서 채워넣는다.

```r
# Data 내용 확인하기
num1 = seq(2, 785, 28) # 시작 2, 끝 785, 등차 28
num2 = seq(29, 785, 28)
count = 0
for(j in 1:28){
  n1 = as.numeric(num1[j])
  n2 = as.numeric(num2[j])
  for(i in seq(n1,n2,1)){
    if(mnist[5,i] != 0){ # mnist의 행값을 바꿔주면 해당하는 행의 값으로 출력된다.
      cat('*')
    }else{
      cat(' ')
    }
    count = count+1
    if(count >27){
      cat('\n')
      count=0
    }
  }
}
```

### 머신러닝

```r
library(nnet)
library(party)
library(randomForest)
library(e1071)
library(caret)
```

```r
# Sampling
samp <- createDataPartition(mnist1$V1, p=0.7, list=F)

# 학습용
data.train <- mnist1[samp,]
nrow(data.train)

# 검증용
data.test <- mnist1[-samp,]

# 문제 및 정답 만들기
x <- subset(data.test, select=-V1)
y <- data.test$V1
```

```r
# Random Forest
  # 학습하기
  rf <- randomForest(V1~.,data=data.train, ntree=100)

  # 검증하기
  pred.rf <- predict(rf, x)

  # 예측율  
  mean(pred.rf == y)
  table(pred.rf,y)
```

```r
  # 학습하기
  ctree <- ctree(V1~.,data=data.train)

  # 검증하기
  pred.ctree <- predict(ctree, x) # 이하, type class 안해도 0과 1로 예측

  # 예측율
  mean(pred.ctree == y)
  table(pred.ctree,y)
```
