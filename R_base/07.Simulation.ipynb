### 동전의 앞면과 뒷면 simulation

```r
x <- runif(1)
if(x <= 1/2){
    result = 0 # 앞면
}else{
    result = 1 # 뒷면
}
print(result)
```

```r
# 알아서 좀 데이터 수집해주라~
coin <- function(n){
    count1 = 0 # 앞면 누적시킬 변수
    count2 = 0 # 뒷면 누적시킬 변수
    for(i in 1:n){
        if(runif(1) <= 1/2){
            count1 = count1 + 1 # 앞면에 1 누적
        }else{
            count2 = count2 + 1 # 뒷면에 1 누적
        }
    }
    cat('앞면 : ', count1,'\n')
    cat('뒷면 : ', count2,'\n')
    cat('--------------------\n')
}
```

```r
coin(10)
coin(1000)
coin(100000)
```

### 가위 바위 보 시뮬

```r
rock <- function(n){
    가위 = 0 # 가위 누적시킬 변수
    바위 = 0 # 바위 누적시킬 변수
    보 = 0 # 보 누적시킬 변수
    for(i in 1:n){
        x=runif(1)
        # 조건문에 runif 넣으면 돌릴때마다 난수 새로 발생.
        if(x <= 1/3){
            가위 = 가위 + 1 # 가위에 1 누적
        }else if(x <= 2/3){
            바위 = 바위 + 1 # 바위에 1 누적
        }else{
            보 = 보 + 1 # 보에 1 누적
        }
    }
    cat('가위 : ', 가위,'\n')
    cat('바위 : ', 바위,'\n')
    cat('보 : ', 보,'\n')
    cat('--------------------\n')
}
```

```r
# 심플하게
rock <- function(n){
    rockR <- rep(0,3)

    for(i in 1:n){
        x=trunc(runif(1)*3)+1
        rockR[x] = rockR[x] + 1 # vector에 알아서 값 삽입된다
    }
    cat('시도 : ',n,'회\n')
    cat('가위 => ',rockR[1],'\n')
    cat('바위 => ',rockR[2],'\n')
    cat('보 => ',rockR[3],'\n')
    cat('--------------------\n')
}
```

```r
rock(10)
rock(1000)
rock(100000)
```

### 주사위 시뮬

```r
dice <- function(n){
    diceR = rep(0,6)
    # diceR 수집
    for(i in 1:n){
        x=trunc(runif(1)*6)+1
        diceR[x] = diceR[x] + 1
    }
    # 출력
    # options(scipen = 999) # 지수표기를 숫자만 출력하도록 변경, 문서 전체 적용
    cat('Throw : ',n,'\n')
    for(j in 1:length(diceR)){
    cat('No',j,'=>',diceR[j],'\n')
    }
    cat('--------------------\n')
}
```

```r
dice(10)
dice(1000)
dice(format(100000,scientific=F)) # '1e+자릿수'의 지수표기를 숫자만 출력하도록 변경
```
