### csv file 한번에 올리기 위한 sql 실행문
SHOW VARIABLES LIKE 'local_infile';   
SET GLOBAL local_infile = 1;

```r
# csv file 읽어서 MySQL에 넣기
data <- read.csv('../data/score.csv', header=F)
colnames(data) <- c('id', 'korean', 'english', 'mathmatics')
data
```

```r
dbWriteTable(conn, 'score', data, overwrite=T, row.names=F) 
# score이라는 테이블에 덮어쓰겠다 data를
# 추가하는 상황이면 append=T 도 포함해야한다.
```

```r
result=dbGetQuery(conn, 'select * from score')
result
```

```r
str(result)
```

```r
data2 <- read.csv('../data/score2.csv', header=F)
colnames(data2) <- c('id', 'korean', 'english', 'mathmatics')
data2
```

```r
dbWriteTable(conn, 'score', data2, overwrite=F, row.names=F, append=T) 
# 추가할 땐 append=T, overwrite=F
result=dbGetQuery(conn, 'select * from score')
result # 11, 12 추가됨
```
