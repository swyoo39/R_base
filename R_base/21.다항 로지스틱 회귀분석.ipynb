```r
# Sampling, 학습용/검증용 만들기
library(caret)
iris2 <- as.data.frame(scale(iris[,-5]))
iris2$Species <- iris$Species
class(iris2)

samp <- createDataPartition(iris2$Species, p=0.7, list=F)
# sample은 index라서 굳이 표준화의 여부에 관계가 없다.
# 학습용
data.tr <- iris2[samp,]
data.tr2 <- iris[samp,]
head(data.tr)
tail(data.tr)
# 검증용
data.test <- iris2[-samp,]
data.test2 <- iris[-samp,]
```

```r
# 문제 및 정답 만들기
x <- subset(data.test, select=-Species)
y <- data.test$Species
x2 <- subset(data.test2, select=-Species)
y2 <- data.test2$Species
```

```r
# 학습하기
model <- multinom(Species~., data=data.tr)
model2 <- multinom(Species~., data=data.tr2)
```

```r
# 검증하기
pred <- predict(model, x)
pred2 <- predict(model2, x2)
```

```r
# 예측율 및 오분류표
mean(pred == y)
table(pred, y)
mean(pred2 == y2)
table(pred2, y2)
```
